```python
import matplotlib.pyplot as plt
import math
def relu(x):
  if(x>0):
    return x;
  else:
    return 0
s = []
x = []
rel=[]
t=0
Tanh=[]
for r in range(0,200):
    x.append(t)
    u = 1/(1+math.exp(-x[r]))
    #print("x = %.3f,\ts = %.3f"%(t,u))
    v = (math.exp(x[r])-math.exp(-x[r]))/(math.exp(x[r])+math.exp(-x[r]))
    Tanh.append(v)
    s.append(u)
    rel.append(relu(x[r]))
    t = t+0.01
plt.subplot(1,2,1)
plt.plot(x,s)
plt.xlabel('x')
plt.ylabel('s')
plt.title('Sigmoid Function')
plt.grid()

plt.subplot(1,2,2)
plt.plot(x,Tanh)
plt.xlabel('x')
plt.ylabel('s')
plt.grid()
plt.title('TanH Function')

plt.figure()
plt.plot(x,rel)
plt.xlabel('x')
plt.ylabel('Relu')
plt.xlim(-1,1)
plt.grid()
plt.title('Relu Function')

plt.show()
```
