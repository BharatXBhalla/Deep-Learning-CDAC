```python
import matplotlib.pyplot as plt
import math
s = []
x = []
t=0
Tanh=[]
for r in range(0,200):
    x.append(t)
    u = 1/(1+math.exp(-x[r]))
    print("x = %.3f,\ts = %.3f"%(t,u))
    v = (math.exp(x[r])-math.exp(-x[r]))/(math.exp(x[r])+math.exp(-x[r]))
    Tanh.append(v)
    s.append(u)
    t = t+0.01
plt.subplot(1,2,1)
plt.plot(x,s)
plt.xlabel('x')
plt.ylabel('s')
plt.title('Sigmoid Function')
plt.grid()

plt.subplot(1,2,2)
plt.plot(x,Tanh)
plt.xlabel('x')
plt.ylabel('s')
plt.grid()
plt.title('TanH Function')

plt.show()
```
